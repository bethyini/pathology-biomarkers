```python
# load the pretrained model
import timm
from timm.data import resolve_data_config
from timm.data.transforms_factory import create_transform
from huggingface_hub import login

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


# pretrained=True needed to load UNI2-h weights (and download weights for the first time)
timm_kwargs = {
            'img_size': 224,
            'patch_size': 14,
            'depth': 24,
            'num_heads': 24,
            'init_values': 1e-5,
            'embed_dim': 1536,
            'mlp_ratio': 2.66667*2,
            'num_classes': 0,
            'no_embed_class': True,
            'mlp_layer': timm.layers.SwiGLUPacked,
            'act_layer': torch.nn.SiLU,
            'reg_tokens': 8,
            'dynamic_img_size': True
        }
model = timm.create_model("hf-hub:MahmoodLab/UNI2-h", pretrained=True, **timm_kwargs)
transform = create_transform(**resolve_data_config(model.pretrained_cfg, model=model))
model = model.to(device)
next(model.parameters()).device  # expected output: device(type='cuda', index=0) if it has been moved to gpu

model.eval()
```

### Extract Tiles (only do this once)

```python
# convert image from svs to jpg, tiling at 224x224 due to large size of svs
# note: there seems to be a lot of empty "whitespace" images, maybe can consider filtering out these images down the line for faster inference
# there are so many jpgs

import openslide
from PIL import Image
import os

def extract_tiles(svs_path, out_dir, tile_size=224, stride=224, level=0):
    os.makedirs(out_dir, exist_ok=True)
    slide = openslide.OpenSlide(svs_path)
    width, height = slide.level_dimensions[level]

    tile_id = 0
    for y in range(0, height, stride):
        for x in range(0, width, stride):
            tile = slide.read_region((x, y), level, (tile_size, tile_size)).convert("RGB")
            tile.save(os.path.join(out_dir, f"tile_{tile_id:05d}.jpg"))
            tile_id += 1

extract_tiles(svs_path, "/content/drive/MyDrive/mlhc/Final project/Train Eval-Elizabeth Kaylee/UNI/data_feed/tiles")
```

```python
!cp -r "/content/drive/MyDrive/mlhc/Final project/Train Eval-Elizabeth Kaylee/UNI/data_feed/tiles" /content/tiles
tile_paths = sorted([os.path.join("/content/tiles", f) for f in os.listdir("/content/tiles")])
```

```python
# extract features for each tile

import numpy as np
from tqdm import tqdm
import os

path = "/content/drive/MyDrive/mlhc/Final project/Train Eval-Elizabeth Kaylee/UNI/data_feed/tiles"
print(os.path.exists(path))          # should be True
print(os.listdir(path)[:5])          # should show your .jpg files
```

```python
features = []

for path in tqdm(tile_paths):
    img = Image.open(path).convert("RGB")
    # img_tensor = transform(img).unsqueeze(0)  # Shape: (1, 3, 224, 224)
    img_tensor = transform(img).unsqueeze(0).to(device)  # makes sure it's on gpu

    with torch.no_grad():
        feat = model(img_tensor)  # Shape: (1, 1536)

    features.append(feat.squeeze(0).cpu().numpy())

features = np.stack(features)  # Final shape: (num_tiles, 1536)
np.save(f"/content/drive/MyDrive/mlhc/Final project/Train Eval-Elizabeth Kaylee/UNI/features/{slide_id}_features.npy", features)
```
