```python
import numpy as np
from skimage.metrics import structural_similarity as ssim
from sklearn.metrics import normalized_mutual_info_score

def compare_heatmaps(hm1, hm2, mask_percentile=90):
    # Ensure same shape
    assert hm1.shape == hm2.shape, "Heatmaps must be same size"
    
    # Flatten for vector metrics
    v1, v2 = hm1.flatten(), hm2.flatten()
    
    # 1) Pearson
    pearson_r = np.corrcoef(v1, v2)[0,1]
    
    # 2) SSIM
    ssim_index, _ = ssim(hm1, hm2, full=True, data_range=1.0)
    
    # 3) Jaccard of top regions
    m1 = hm1 > np.percentile(hm1, mask_percentile)
    m2 = hm2 > np.percentile(hm2, mask_percentile)
    jaccard = np.logical_and(m1,m2).sum() / np.logical_or(m1,m2).sum()
    
    # 4) Mutual Information
    nmi = normalized_mutual_info_score(v1, v2)
    
    return {
        "pearson_r": pearson_r,
        "ssim": ssim_index,
        f"jaccard_top_{100-mask_percentile}%": jaccard,
        "nmi": nmi
    }
```

```python
hm1 = np.load("CLAM/heatmap_outputs/BRAF/arrays/TCGA-A6-2672_smoothed_heatmap.npy")
hm2 = np.load("CLAM/heatmap_outputs/MSI/arrays/TCGA-A6-2672_smoothed_heatmap.npy")

# Example usage:
metrics = compare_heatmaps(hm1, hm2, mask_percentile=90)
for k,v in metrics.items():
    print(f"{k}: {v:.3f}")
```

```python
import matplotlib.pyplot as plt

diff = np.abs(hm1 - hm2)
plt.imshow(diff, cmap='viridis')
plt.title('Absolute difference map')
plt.colorbar()
plt.show()
```

```python
# look at all 14 heatmaps and get the average

import numpy as np
import os
import pandas as pd

# your list of 14 slideIDs
slide_ids = ['TCGA-AA-3947', 'TCGA-AA-3833', 'TCGA-AA-A01P', 'TCGA-AA-3543', 'TCGA-AA-3525', 'TCGA-AA-3821', 
             'TCGA-AA-3715', 'TCGA-AA-3877', 'TCGA-AA-3516', 'TCGA-AA-3966', 'TCGA-AA-3949', 'TCGA-A6-2672', 'TCGA-AA-A022', 'TCGA-AA-A00J']

results = []

for sid in slide_ids:
    # paths to your saved .npy files
    path1 = f"CLAM/heatmap_outputs/BRAF/arrays/{sid}_smoothed_heatmap.npy"
    path2 = f"CLAM/heatmap_outputs/MSI/arrays/{sid}_smoothed_heatmap.npy"
    if not (os.path.exists(path1) and os.path.exists(path2)):
        print(f"Skipping {sid}, file(s) missing.")
        continue

    hm1 = np.load(path1)
    hm2 = np.load(path2)
    m   = compare_heatmaps(hm1, hm2, mask_percentile=90)
    m["slide_id"] = sid
    results.append(m)

# turn into DataFrame
df = pd.DataFrame(results)

# compute mean (and optionally std)
mean_metrics = df.drop(columns="slide_id").mean()
std_metrics  = df.drop(columns="slide_id").std()

print("Average across slides:")
for k,v in mean_metrics.items():
    print(f"  {k}: {v:.3f}  {std_metrics[k]:.3f}")
```

```python
csv_path = "heatmap_similarity_metrics.csv"
df.to_csv(csv_path, index=False)
print(f"Metrics saved to {csv_path}")
```
